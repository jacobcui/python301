# Summarising, Aggregating, and Grouping data

```python
# Load csv using pandas.read_csv()
# Load excel using pandas.read_excel()

import pandas as pd
excel_file = 'xls_sample/sales.xlsx'
df = pd.read_excel(excel_file)
```

```python
# how many rows in the data set
df['Account'].count()
```

```python
# What is the largest amount of price
df['Price'].max()
```

```python
# Which row has the largest amount of price
dir(df['Price'].max())
df['Price'].idxmax()
```

```python
# How much is the total sales
df['Price'].sum()
```

```python
# How much is manager Debra Henley's sales
df['Price'][df['Manager'] == 'Debra Henley'].sum()
```

```python
# How many entries are for each representative.
df['Rep'].value_counts()
```

```python
# How many of non-null unique client names.
df['Name'].nunique()
```

```python
# How many representatives.
df['Rep'].nunique()
```

### All other [basic statistics functions reference](http://pandas.pydata.org/pandas-docs/stable/getting_started/basics.html)


```python
%matplotlib inline
df.plot.bar(x='Rep', rot=60)
```

```python
# Summarising Groups
# The groupby() function returns a GroupBy object, 
# but essentially describes how the rows of the original data set has been split.
df.groupby('Product').groups
```

```python
# Get the first entry for each Product
df.groupby('Product').first()
```

```python
# Get the sum of sales per Product.
df.groupby('Product')['Price'].sum()
```

```python
# Get the sum of sales per sales representative.
df.groupby('Rep')['Price'].sum()
```

```python
# What is the sum of sales for CPU for each representative.
df[df['Product'] == 'CPU'].groupby('Rep').sum()
```

You can also group by more than one variable, allowing more complex queries.


```python
# List representative's sales by manager.
df.groupby(['Manager', 'Rep'])['Price'].sum()
```

Functions like max(), min(), mean(), first(), last() can be quickly applied to the GroupBy object to obtain summary statistics for each group

```python
# List representative's sales per client.
df.groupby(['Manager', 'Rep', 'Name'])['Price'].sum()
```

### Multiple Statistics per Group
agg() function  

Instructions for aggregation are provided in the form of a python dictionary or list.  

The dictionary keys are used to specify the columns upon which youd like to perform operations, and the dictionary values to specify the function to run.

```python
df.groupby(['Manager', 'Rep', 'Product']).agg(
    {'Price': 'sum',     # find the sum of the Price for each group
     'Name': 'count',    # find the number of clients
     'Status': 'first'   # get the first status per group
    })
```

```python
# Applying multiple functions to columns in groups
df.groupby(['Manager', 'Rep', 'Product']).agg(
    {'Price': 'sum',     # find the sum of the Price for each group
     'Name': 'count',    # find the number of clients
     'Status': ['first', 'last']   # get the first status per group
    })
```
